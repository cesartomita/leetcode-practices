```python
import pandas as pd

data = {
    "patient_id": [1, 2, 3, 4, 5, 6],
    "patient_name": ["Daniel", "Alice", "Bob", "George", "Alain", "Maria"],
    "conditions": [
        "SADIAB100",
        "",
        "DIAB100 MYOP",
        "ACNE DIAB100",
        "DIAB201",
        "DIAB101 MYOP"
    ]
}

patients = pd.DataFrame(data)

patients
```

```python
patients_with_diabetes = patients[patients["conditions"].str.contains("DIAB1", regex=False)]

patients_with_diabetes
```

```python
patients_with_diabetes_split = patients.copy()

patients_with_diabetes_split["split_conditions"] = patients["conditions"].str.split(" ")

patients_with_diabetes_split
```

```python
result = patients_with_diabetes_split[
    patients_with_diabetes_split["split_conditions"].apply(lambda condition: "DIAB100" in condition)
]

result
```

```python
patients_with_diabetes_split_exploded = patients_with_diabetes_split.explode("split_conditions") # explode() é um método que transforma cada elemento de uma lista dentro de uma célula em uma linha separada.

patients_with_diabetes_split_exploded
```

```python
exploded_filter = patients_with_diabetes_split_exploded["split_conditions"].str.startswith("DIAB1", na=False) # na=False faz o método retornar False para células NaN, evitando erros no filtro

exploded_filter
```

```python
patients_with_diabetes_split_exploded[exploded_filter]
```
